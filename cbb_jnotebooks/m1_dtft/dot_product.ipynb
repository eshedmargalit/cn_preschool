```matlab
%% DOT PRODUCT AND CONVOLUTION %%
% quick definition: element wise multiplication between two vectors of equal length, then add them up.
```

```matlab
% establish vectors
x = 1:10;
y = rand(1,10);
```

```matlab
% make vectors equal in length
a = x / norm(x);
b = y / norm(y);
```

```matlab
% signal processing - sum of a weighted by elements of b
element_wise_multiplication = a .* b; %[a(1)*b(1) a(2)*b(2) a(3)*b(3) ... a(10)*b(10) ]
dot_ab = sum(element_wise_multiplication)
```

```matlab
% geometric part I
r = [-25,25];
g = [-25,-25];
c = [10,5];
gr = r / norm(r);
gg = g / norm(g);
gc = c / norm(c);
[fh,ah] = makefigure();
lh_gr = line([0, gr(1)], [0, gr(2)], 'color', 'r');
lh_gg = line([0, gg(1)], [0, gg(2)], 'color', 'g');
lh_gc = line([0, gc(1)], [0, gc(2)], 'color', 'c');
legend(ah,[lh_gr, lh_gg, lh_gc], {'gr', 'gg', 'gc'})

% forking render corrrectly
lh_x = line([-2,2], [0,0],'color', 'k');
lh_y = line([0, 0], [-2, 2], 'color', 'k');
set(fh,'paperposition', [0,0,7,7])
axis(ah,[-1 1 -1 1])
```

```matlab
% geometric part II
[fh,ah] = makefigure();
lh_gr = line([0, gr(1)], [0, gr(2)], 'color', 'r');
lh_gg = line([0, gg(1)], [0, gg(2)], 'color', 'b');
lh_gc = line([0, gc(1)], [0, gc(2)], 'color', 'c');
legend(ah,[lh_gr, lh_gg, lh_gc], {'gr', 'gb', 'gc'})

% plot the dot product proj
drg = dot(gr,gg)
dgc = dot(gg,gc)
drc = dot(gr,gc)
lh_drc_proj_gr = line([0, drc * gr(1)], [0, drc * gr(2)], 'color', 'r','linewidth',2,'parent',ah);
lh_dgc_proj_gr = line([0, dgc * gg(1)], [0, dgc * gg(2)], 'color', 'g','linewidth',2,'parent',ah);

% show right angles - pretend gc is the new zero
cr_ra = line([gc(1),gc(1)+gr(1)], [gc(2),gc(2)-gr(2)], 'linestyle','--', 'color','r');
rb_ra = line([gc(1),gc(1)-gg(1)], [gc(2),gc(2)+gg(2)], 'linestyle','--', 'color','g');
cr_ra = line([gc(1),gc(1)-gr(1)], [gc(2),gc(2)+gr(2)], 'linestyle','--', 'color','r');
rb_ra = line([gc(1),gc(1)+gg(1)], [gc(2),gc(2)-gg(2)], 'linestyle','--', 'color','g');

% forking render corrrectly
lh_x = line([-2,2], [0,0],'color', 'k');
lh_y = line([0, 0], [-2, 2], 'color', 'k');
set(fh,'paperposition', [0,0,7,7]); % so angles look right
axis(ah,[-1 1 -1 1])

```

```matlab
% covariance between two vectors
avg_a = mean(a); % E[x]
avg_b = mean(b); % E[y]
% demean the vectors -- (x - E[x])
avg_sub_a = a - avg_a;
avg_sub_b = b - avg_b;
% cov -- E[(x - E[x])(y - E[y])]
covar = mean(avg_sub_a/norm(avg_sub_a) .* avg_sub_b/norm(avg_sub_b))
% dot product divided by the elements of avg_sub_a .* avg_sub_b
dot_scaled = dot(avg_sub_a/norm(avg_sub_a),avg_sub_b/norm(avg_sub_b))/numel(avg_sub_a)
```
